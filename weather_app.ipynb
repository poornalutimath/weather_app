```python
import requests
import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk
import io

API_KEY = 'a05b6b6be725845af7e1a568fba6f209'  # Replace with your OpenWeatherMap API key
BASE_URL = 'http://api.openweathermap.org/data/2.5/weather'

def get_weather(city):
    params = {'q': city, 'appid': API_KEY, 'units': 'metric'}
    response = requests.get(BASE_URL, params=params)
    return response.json()

def display_weather(data):
    if data['cod'] == 200:
        city = data['name']
        temp = data['main']['temp']
        humidity = data['main']['humidity']
        weather_description = data['weather'][0]['description']
        icon_code = data['weather'][0]['icon']
        weather_icon = get_weather_icon(icon_code)
        return city, temp, humidity, weather_description, weather_icon
    else:
        return None

def get_weather_icon(icon_code):
    url = f"http://openweathermap.org/img/wn/{icon_code}.png"
    response = requests.get(url)
    image_data = response.content
    image = Image.open(io.BytesIO(image_data))
    return ImageTk.PhotoImage(image)

def on_search():
    city = entry_city.get()
    weather_data = get_weather(city)
    if weather_data:
        city, temp, humidity, weather_description, weather_icon = display_weather(weather_data)
        label_city.config(text=f"City: {city}")
        label_temp.config(text=f"Temperature: {temp}°C")
        label_humidity.config(text=f"Humidity: {humidity}%")
        label_description.config(text=f"Weather: {weather_description}")
        label_icon.config(image=weather_icon)
        label_icon.image = weather_icon
    else:
        messagebox.showerror("Error", "City not found. Please check the name and try again.")

def create_gui():
    window = tk.Tk()
    window.title("Weather App")

    global entry_city, label_city, label_temp, label_humidity, label_description, label_icon

    tk.Label(window, text="Enter city name:").grid(row=0, column=0)
    entry_city = tk.Entry(window)
    entry_city.grid(row=0, column=1)
    tk.Button(window, text="Search", command=on_search).grid(row=0, column=2)

    label_city = tk.Label(window, text="City:")
    label_city.grid(row=1, columnspan=3)

    label_temp = tk.Label(window, text="Temperature:")
    label_temp.grid(row=2, columnspan=3)

    label_humidity = tk.Label(window, text="Humidity:")
    label_humidity.grid(row=3, columnspan=3)

    label_description = tk.Label(window, text="Weather:")
    label_description.grid(row=4, columnspan=3)

    label_icon = tk.Label(window)
    label_icon.grid(row=5, columnspan=3)

    window.mainloop()

if __name__ == "__main__":
    create_gui()
```
